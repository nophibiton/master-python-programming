```python
# integrate a for-loop with if-statement
# the percent operator % is the remainder after division

for n in range(0,18):
  if n%2 == 0:
    print(f"{n} is an even number.")
  else:
    print(f"{n} is an odd number. ")
```

```python
# create a function that reports the fibonacci series for n+3 numbers
#
# 0, 1, 1, 2, 3, 5, 8
#
```

```python
def fib_series(x):
  v = [0,1,1]
  for i in range(x):
    v.append(sum(v[-2:]))
  return v
```
