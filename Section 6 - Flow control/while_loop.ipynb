```python
# while loops

toggle = True
i = 0

while toggle:
  print(i)
  i+=1
  if i==8:
    toggle=False
```

```python
i = 0
while True:
  print(i)
  i+=1
  if i==8:
    break
```

```python
# Exercise
# make a Poisson-like counter
#
# start a parameter lambda. positive number (10)
# start at 1, multiply by random number uniform([0,1])
# keep drawing until being smaller than exp(-lambda)
# the number of iterations down until target is <exp(-lambda)

import numpy as np

def PoissonCounter(lambda_param):
  lambda_param = 10.0
  ctr,num = 0,1 
  while num > np.exp(-lambda_param):
    ctr += 1
    num *= np.random.uniform(low=0,high=1.0)
  return ctr
```

```python
p = np.zeros(100)
for i in range(0,100):
  p[i] = PoissonCounter(10)

p
```
