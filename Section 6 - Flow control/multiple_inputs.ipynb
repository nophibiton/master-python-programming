```python
# Exercise
# Create a dictionary out of these two list
#
names = ['alpha','beta','gamma']
values = [10,20,40]

my_dict = dict()
for i,j in zip(names,values):
  my_dict[i] = j

print(my_dict)
```

```python
# alternative solution
names = ['alpha','beta','gamma']
values = [10,20,45]

my_dict = dict(zip(names,values))

print(my_dict)
```
