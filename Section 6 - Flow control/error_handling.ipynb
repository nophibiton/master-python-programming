```python
# create a function
# input 2 numbers and output their product
def funfun(x,y):
  # check inputs if numeric
  if not isinstance(x,(int,float)):
    # check if x can be converted into a number
    if isinstance(x,str) and x.isnumeric():
      print(f"Note: Converted '{x}' to float.")
      x = float(x)
    else:
      raise Exception('Input x must be a number!')
  
  if not isinstance(y,(int,float)):
    raise Exception('Input y must be a number!')

  z = x*y
  return z

funfun('a2',3)
```
