An `else` statement after the `if` statement is used/implemented when the `if` statement is false.

```python
r = np.random.randn()

if r>0:
  print(f"{r} is positive.")
else:
  print(f"{r} is non-positive.")
```

Another flow control operator is the `elif` which when the first `if` statement is false, another conditional statement is required/asked.

```python
caketype = ''

if caketype == 'carrot':
  print('Total yumminies!')
elif caketype == 'chocolate':
  print('Still quite good.')
elif caketype == 'fruit':
  print('Healthy??')
else:
  print('Cake type not recognized.')
```

```python
# one-line if statements
caketype = 'Carrot'

if caketype[0].lower() == 'c': print('Probably carrot cake.')

print('Probably carrot cake') if caketype[0].lower() == 'c' else print ('Or not.')

```

```python
# conjunctive conditional
if 1<3 and 4==4:
  print('Both are true.')
```

```python
# Exercise
# Create a dot product function
# function checks that both inputs are numpy arrays (hit: isinstance)
# function checks for same length of two input vectors
# returns dot product if pass, usedful error message otherwise (hint: raise)
```

```python
def check_np_arrays(array1,array2):

  check1 = isinstance(array1,np.ndarray)
  check2 = isinstance(array2,np.ndarray)
  if check1==True and check2==True:
    check_arrays = True
  else:
    check_arrays = False

  return check_arrays
```

```python
def check_same_length(array1,array2):

  if len(array1) == len(array2):
    check = True
  else:
    check = False

  return check
```

```python
def my_dot_product(array1,array2):

  if check_np_arrays(array1,array2) == True and check_same_length(array1,array2) == True:
    # calculate dot product here
    s = len(array1)
    dot_prod = 0
    for i in range(0,s):
      prod = array1[i] * array2[i]
      dot_prod += prod
  else:
    raise Exception('Unable to calculate the dot product. Check whether the input is a numpy array and have the same lengt')

  return dot_prod


```
