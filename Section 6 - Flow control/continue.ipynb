A way to skip line in a `for` loop or a `while` loop.

```python
for i in range(0,10):
  if i%2 == 0:
    continue
  print(i)
```

```python
text = 'HellO, mu name is MikE'

for i in text:
  if i in 'aeiou':
    continue
  print(i)
```

```python
text = 'HellO, my name is MikE'

for i in text:
  if i in 'aeiou' or i in 'AEIOU':
    continue
  print(i)
```

```python
# Another solution
text = 'HellO, my name is MikE'

for i in text:
  if i.lower() in 'aeiou':
    continue
  print(i)
```
