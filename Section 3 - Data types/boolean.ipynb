```python
# Exercise
# Ask the user for Pythagorean triplets, and test whether that's a real triplet
# example 
# 3, 4, 5
# 5, 12, 13

print('Enter three numbers and I will tell if it is a Pythagorean triplet')
xa = float(input('Enter first number: '))
xb = float(input('Enter second number: '))
xc = float(input('Enter third number: '))


checkPyth = (xc**2 == xa**2 + xb**2) or (xa**2 == xb**2 + xc**2) or (xb**2 == xa**2 + xc**2)

print(f"It is {checkPyth} Pythagorean Triplet.")

```
