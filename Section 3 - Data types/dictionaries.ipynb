```python
# Dictionary is simillar to structure with Matlab
D = dict()
D
```

```python
# key/value pair
# key = name, value = Mike
D['name'] = 'Mike'
# key = AgeRange, value 20-50
D['AgeRange'] = [20,50]

D
```

```python
# another way of making a dictionary using {}
E = {'name':'Nophi','AgeRange':[20,50]}
E
```

```python
# to retrieve a value in dictionary
E['name']
```

```python
# use the methods keys()
E.keys()
```

```python
# keys in dictionaries are case sensitive
'Nophi' in E # only searching for keys
```

```python
E.values()
```

```python
E.items()
```
