The difference of a Tuple with a List: a Tuple is immutable while a List is mutable.


*   You can not change an element in a tuple.
*   You can however change the entire tuple.



```python
# Tuples
aTuple = (3,'4',3)

# List
aList = [3,'4',3]

# tuples are immutable (it is fixed, cannot replace the items)
# lists are mutable

print(aTuple)
print(aList)
aList[0] = 5

#aTuple = [5.,'4',3]


print(aTuple)
print(aList)
```
