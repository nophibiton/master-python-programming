```python
def funfun(in1,in2):
  sumres = in1+in2
  prodres = in1*in2
  print(f"Their sum is {sumres}.")
  print(f"Their product is {prodres}.")
  return sumres, prodres
```



*   If a function returns multiple output and the it save into a single variable then it will saved as a tuple.
*   The output can also be assigned to different variables.



```python
out = funfun(4,4)
print(out)
type(out)
outS,outP = funfun(4,4)
print(outS)
type(outS)

```

```python
# Exercise 
# Create a function that computes a factorial of an integer
# 5! = 5*4*3*2*1
def myfactorial(num):
  alist = np.array(list(range(1,num+1)))
  fact = alist.prod()
  return fact
```

```python
import math

#help(math.factorial)
val = 23
math_fact = math.factorial(val)
my_fact = myfactorial(val)

print(f"The factorial of {val} using math module is {math_fact} while using my factorial function is {my_fact}")
```
