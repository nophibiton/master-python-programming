```python
# OOP (classes)

class Experiment(object):
  # constructor method
  def __init__ (self,eType,dataN, status='ongoing'):
    self.expType = eType
    self.dataCount = dataN
    self.expStatus = status

  def increaseN(self,number):
    self.dataCount += number

  def changeExpStatus(self,status):
    self.expStatus = status

  def __str__(self):
    return 'Experiment type: ' + self.expType + ' has ' + str(self.dataCount) + ' data points.'
```

```python
myexperiment = Experiment('Brain',1450)

print(myexperiment.dataCount)
print(myexperiment.expType)
print(myexperiment.expStatus)
myexperiment.increaseN(10)
print(myexperiment.dataCount)

print(myexperiment)
```

```python
class boardgame(object):
  def __init__(self,name,numPlayed,numWon):
    self.name = name
    self.numPlayed = numPlayed
    self.numWon = numWon

  def updateWins(self,nwins):
    self.numWon += nwins

  def updatePlays(self,nplays):
    self.numPlayed += nplays

    
```

```python
myboardgame = boardgame('game of the generals',10,8)
print(myboardgame.name)
print(myboardgame.numPlayed)
print(myboardgame.numWon)

myboardgame.updateWins(2)
myboardgame.updatePlays(5)

print(myboardgame.numPlayed)
print(myboardgame.numWon)
```
