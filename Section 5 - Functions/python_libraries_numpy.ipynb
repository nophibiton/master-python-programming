Python is not usually built with complex mathametical functions. A module can be imported to the python environment to use math functions.

```python
lst = [10,20,30]
```

numpy is a module used for scientific computing.

```python
import numpy as np # numpy is given an alias as 'np'

np.mean(lst) # start with np the dot to use the functions in numpy
```

You only need to import once the module (i.e. numpy). Usually the module is imported at the top of the script.

```python
np.linspace(1,80,20)
```

```python
number = [1,2,3,4,5,6]
numbernp = np.array(number)
type(number)
type(numbernp)

print(number)
print(numbernp)
```

A method is like a function but is attached to a particular object or variable.

```python
numbernp.max()
# alternatively it can be defined as 
np.min(numbernp)
```

Some methods may change the variable itself with explicitly assigning them.

```python
n = [4,6,7,3,4,5,7,9,1]
print(n)
n.sort() # the variable n is changed with sorted version of its elements
print(n)
```

Sometimes applying function can change a variable type.

```python
n = [4,6,7,3,4,5,7,9,1]
print(type(n))
n = np.sort(n)
print(type(n))
```
