The `x` and `y` variables have local scope. That is, it is only usable with the function `funfun`.

```python
def funfun():
  x = 7
  y = 10
  print(x)
```

```python
funfun()
```

```python
x = 3
funfun()
print(x)
print(y)
```

```python
def fz():
  print(z)
```

```python
z = 3
fz()
```

```python
# Rules
# 1) variables created inside a function are local (accessible only inside the function)
# 2) variables created outsied a function are global (accessible in or out of the function)
```

```python
# Exercise
# write a function that flips a coin N times and reports the average.
import numpy as np

def flipCoin(N):
  ave_flip = np.mean (np.random.randint(low=0,high=2,size=N))
  return ave_flip

flipCoin(100000000)

```
