```python
import numpy as np
import matplotlib.pyplot as plt


N = 100000
x = np.zeros(N)
y = np.zeros(N)

for i in range(1,N):
  k = np.random.randint(1,4)

  x[i] = 0.5*x[i-1] + k-1
  y[i] = 0.5*y[i-1]

  if k==2: y[i] += 2

fig,ax = plt.subplots()

ax.plot(x,y,'k.',markersize=10)
plt.axis('off')
plt.show()
```
