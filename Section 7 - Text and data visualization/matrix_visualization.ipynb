```python
import matplotlib.pyplot as plt
import numpy as np

m = 3
n = 5

M = np.random.randint(10,size=(m,n))
print(M)


plt.imshow(M)

for i in range(0,m):
  for j in range(0,n):
    plt.text(j,i,str(M[i,j]),horizontalalignment='center',
             fontsize=30, verticalalignment='center')

plt.colorbar()
plt.show()
```

```python
# Exercise:
# 1) Create a hilbert matrix (10x10) H[i,j] = 1 / (i+j-1)
# 2) visualize the Hilbert matrix using seaborn? Hint sns.heatmap

import numpy as np
import seaborn as sns

# create hilbert matrix
n = 10
h_matrix = np.zeros((n,n))

for i in range(0,h_matrix.shape[0]):
  for j in range(0,h_matrix.shape[1]):
    n,m = i+1,j+1
    h_matrix[i,j] = 1/(n+m-1)

#np.round(h_matrix,2)

# visualize the matrix
sns.heatmap(h_matrix,vmin=0,vmax=0.3)
plt.show()
```
