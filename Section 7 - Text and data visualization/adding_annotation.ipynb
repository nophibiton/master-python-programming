```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
x = np.arange(-5,6)
y = x**2

fig,ax = plt.subplots()

ax.plot(x,y,'ko-',markerfacecolor='w',markersize=15)
ax.annotate('Hello',(x[3],y[3]),fontsize=15,arrowprops=dict(),
            xycoords='data',xytext=(0,10),horizontalalignment='center')


plt.show()
```

```python
#
# Exercise: annotate the minimum
#  
x = np.arange(-5,6)
y = x**2 + np.cos(x)*10
idx_min = np.where(y == np.amin(y))

fig,ax = plt.subplots()

ax.plot(x,y,'ko-',markersize=15,markerfacecolor='m')

#min_val = 'min: ('+ str(x[idx_min[0][0]])+','+ str(round(y[idx_min[0][0]],3)) + ')'
min_val = f'min: ({x[idx_min[0][0]]},{y[idx_min[0][0]]:.2f})'

ax.annotate(min_val,(x[idx_min[0][0]],y[idx_min[0][0]]), fontsize=15,
            arrowprops=dict(arrowstyle ='wedge',color='m'),xycoords='data',
            xytext=(x[int(len(x)/2)],np.max(y)*.8),horizontalalignment='center')


ax.set_xlabel('x')
ax.set_ylabel('y=x^2+10cos(x)')
plt.show()
```
