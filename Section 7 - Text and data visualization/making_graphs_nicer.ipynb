```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-3,3,101)

plt.plot(x,x,label='y=x')
plt.plot(x,x**2,label='y=x**2')
plt.plot(x,x**3,label='y=x**3')

plt.legend()

plt.show()
```

```python
x = np.linspace(-3,3,101)

plt.plot(x,x)
plt.plot(x,x**2)
plt.plot(x,x**3)

plt.legend(['y=x','y=x**2','y=x**3'])

plt.show()
```

```python
x = np.linspace(-3,3,101)

plt.plot(x,x,label='y=x')
plt.plot(x,x**2,label='y=x**2')
plt.plot(x,x**3,label='y=x**3')

plt.xlabel('x')
plt.ylabel('y=f(x)')

plt.xlim([x.min(),x.max()])
plt.ylim([-10,10])

#plt.gca().set_aspect('equal')
#plt.gca().set_aspect(0.3)
plt.gca().set_aspect(1/plt.gca().get_data_ratio())
plt.title('Polynomial Functions')
plt.legend()
plt.show()
```

Plot commands used above slight differs when there are subplots.

```python
fig,ax=plt.subplots()

ax.plot(x,x,'r',label='x')
ax.plot(x,x**2,color=[0,1,0],label='x**2')
ax.plot(x,x**3,color=[0,0,0],label='x**3')

ax.set_xlabel('x')
ax.set_ylabel('y=f(x)')
ax.set_title('Polynomial Functions')

ax.set_xlim([x.min(),x.max()])
ax.set_ylim([-10,10])

ax.set_aspect(1/ax.get_data_ratio())

ax.legend()
ax.grid()
plt.show()
```
