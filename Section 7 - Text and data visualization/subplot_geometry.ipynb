```python
import matplotlib.pyplot as plt
import numpy as np


plt.subplot(2,1,1)
plt.plot(np.random.randn(10))

plt.subplot(2,1,2)
plt.plot(np.random.randn(10))

plt.show()
```

```python
for i in range(0,9):
  plt.subplot(3,3,i+1)
  plt.plot(np.random.randn(7))
plt.show()
```

```python
fig,ax = plt.subplots(1,3,figsize=(9,3)) # different command

x = np.arange(10)
ax[0].plot(x,x**2,'b')
ax[1].plot(x,np.sqrt(x),'r')
ax[2].plot(x,x,'g')

plt.show()
```

```python
fig,ax = plt.subplots(2,2)

ax[0,0].plot(np.random.randn(4,4))
ax[0,1].plot(np.random.randn(4,4))
ax[1,0].plot(np.random.randn(4,4))
ax[1,1].plot(np.random.randn(4,4))
plt.show()
```

```python
# Exercise: create a 3x3 subplot geometry
# populate the subplots using a for loop
# np.flatten()
fig,ax = plt.subplots(3,3)

for i in range(0,ax.shape[0]):
  for j in range(0,ax.shape[1]):
    ax[i,j].plot(np.random.randn(4,4))

plt.show()
```

```python
# alternative solution
fig,ax = plt.subplots(3,3,figsize=(8,8))

for i in ax.flatten():
  i.plot(np.random.randn(2,8))

plt.show()
```
